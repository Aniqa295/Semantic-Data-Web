```python
import import_ipynb
import dash
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_bootstrap_components as dbc
import dash_html_components as html
import dash_table
from dash.dependencies import Input, Output, State

from ipynb.fs.full.MainApp import app
from ipynb.fs.full.HelperFunctions import df
from ipynb.fs.full.HelperFunctions import start_table_df
```

```python
#-----------------------------------------selected tab callback ----------------------------------------------------------------
#render page for selected tab
@app.callback(
    Output('tabs-content', 'children'),
    Input('tabs', 'value'),
)
def render_content(tab):
    if tab == 'Dashboard':
        return html.Div([
            html.Div(id='output-data-upload'),
            html.Div(id='output-data-upload2'),
            html.Div([
                html.H3('Statistics'),
                html.Div(id='records-size',children=[]),
                html.Div(id='records-size2',children=[])
            ])
        ])
    elif tab == 'Tables':
        return html.Div([
            html.Div(
                id='MainTableDiv',
                children=[
                    dash_table.DataTable(
                        id='Resulttable',
                        style_table={'height': '250px', 'overflowY': 'auto'},
                        data=start_table_df.to_dict('records'), 
                        columns = [{'id': c, 'name': c} for c in start_table_df.columns],
                        style_cell=dict(textAlign='left'),   
                        editable=True,
                        filter_action="native",
                        sort_action="native",
                        sort_mode="multi",
                        column_selectable="single",
                        row_selectable="multi",
                        row_deletable=True,
                        selected_columns=[],
                        selected_rows=[],
                        page_action="native",
                        page_current= 0,
                        page_size= 10,
                        export_format="csv"
                    )
                ],
                style= {
                    'display': 'none',
                    'marginTop': '5px'
                }
            ),
            html.Div([
                 dbc.Alert(
                     ["There are no data for generating table."],
                     id="alert-table",
                     is_open=False,
                     fade=True,
                     dismissable=True,

                 ),    
            ]),
            html.Div(
                id='MainTableDiv2',
                children=[
                    dash_table.DataTable(
                        id='Resulttable2',
                        style_table={'height': '250px', 'overflowY': 'auto'},
                        data=start_table_df.to_dict('records'), 
                        columns = [{'id': c, 'name': c} for c in start_table_df.columns],
                        style_cell=dict(textAlign='left'),   
                        editable=True,
                        filter_action="native",
                        sort_action="native",
                        sort_mode="multi",
                        column_selectable="single",
                        row_selectable="multi",
                        row_deletable=True,
                        selected_columns=[],
                        selected_rows=[],
                        page_action="native",
                        page_current= 0,
                        page_size= 10,
                        export_format="csv"
                    )
                ],
                style= {
                    'display': 'none',
                    'marginTop': '5px'
                }
            ),
            
            html.Div([
                 dbc.Alert(
                     ["There are no data for generating table."],
                     id="alert-table2",
                     is_open=False,
                     dismissable=True,
                     style={
                            'position': 'absolute',
                            'top': '15%', 
                            'right': '10%', 
                            'width': '40%',
                            'height': '5%',
                            'lineHeight': '20px',
                            'borderWidth': '1px',
                            'borderStyle': 'groove',
                            'margin': 'auto',
                            'fontSize': 'small',
                            'color': 'red'
                            },
                         
                    
                 ),    
                    ]),
            
        ])
    elif tab == 'Charts':
        return html.Div([
               html.Div(
                   id="result-graph",
                   className="result-graph",
                   children=[
                        dcc.Graph(
                            id='graph',
                            figure={
                                "layout": {
                                    "height": 253,
                                    'overflow': 'scroll'
                                },
                            },
                        ),
                
                     ],
                 style= {'display': 'none'}  
               ),

                html.Div([
                     dbc.Alert(
                         ["There are not enough data for generating chart."],
                         id="alert-chart",
                         is_open=False,
                         fade=True,
                         dismissable=True,
                         style={
                            'position': 'absolute',
                            'top': '15%', 
                            'right': '10%', 
                            'width': '40%',
                            'height': '5%',
                            'lineHeight': '20px',
                            'borderWidth': '1px',
                            'borderStyle': 'groove',
                            'margin': 'auto',
                            'fontSize': 'small',
                            'color': 'red'
                            },
                         
                        
                        
                     ),    
                 ]),
    
                html.Div(
                       id="result-graph2",
                       className="result-graph",
                       children=[
                            dcc.Graph(
                                id='graph2',
                                figure={
                                "layout": {
                                    "height": 270, 
                                },
                            },
                            ),

                         ],
                     style= {'display': 'none'}  
                   ),

                html.Div([
                     dbc.Alert(
                         ["There are not enough data for generating chart."],
                         id="alert-chart2",
                         is_open=False,
                         dismissable=True,
                         style={
                            'position': 'absolute',
                            'top': '15%', 
                            'right': '10%', 
                            'width': '40%',
                            'height': '5%',
                            'lineHeight': '20px',
                            'borderWidth': '1px',
                            'borderStyle': 'groove',
                            'margin': 'auto',
                            'fontSize': 'small',
                            'color': 'red'
                            },
                         
                     )   
                 ]),

        ])
    elif tab == 'Maps':
        return html.Div([
               html.Div(
                   id="result-map",
                   className="result-map",
                   children=[
                        dcc.Graph(
                            id='map',
                            figure={
                                #"layout": {
                                 #   "height": 500, 
                                    
                                #},
                            },
                        ),                
                     ],
                 style= {'display': 'none'}  
               ),
               html.Div([
                    dbc.Alert(
                    ["There are no coordinates for generating map."],
                    id="alert-map",
                    is_open=False,
                    fade=True,
                    dismissable=True,
                    style={
                            'position': 'absolute',
                            'top': '15%', 
                            'right': '10%', 
                            'width': '40%',
                            'height': '5%',
                            'lineHeight': '20px',
                            'borderWidth': '1px',
                            'borderStyle': 'groove',
                            'margin': 'auto',
                            'fontSize': 'small',
                            'color': 'red'
                            },
                         
                   
                    ),
               ]),
            
               html.Div(
                   id="result-map2",
                   className="result-map",
                   children=[
                        dcc.Graph(
                            id='map2',
                            figure={
                                #"layout": {
                                #    "height": 500, 
                                #},
                            },
                        ),

                     ],
                 style= {'display': 'none'}  
               ),
               html.Div([
                    dbc.Alert(
                    ["There are no coordinates for generating map."],
                    id="alert-map2",
                    is_open=False,
                    fade=True,
                    dismissable=True,
                    style={
                            'position': 'absolute',
                            'top': '15%', 
                            'right': '10%', 
                            'width': '40%',
                            'height': '5%',
                            'lineHeight': '20px',
                            'borderWidth': '1px',
                            'borderStyle': 'groove',
                            'margin': 'auto',
                            'fontSize': 'small',
                            'color': 'red'
                            },
                         
                    
                        )
               ])
        ])
    elif tab == 'Samples':
        return html.Div([
            html.Div(
                className="examples",
                children=[
                    dash_table.DataTable(
                        id='querySamples',
                        columns=[{"name": i, "id": i} for i in df.columns],
                        style_data_conditional=[
                            {
                                'if': {
                                    'state': 'active'  # 'active' | 'selected'
                                },
                                'backgroundColor': 'rgb(211, 224, 234)',
                                'border': '1px solid rgb(0, 116, 217)'
                            }
                        ],
                        data=df.to_dict('records'),
                        style_table={'width': '800px'},
                        style_cell=dict(textAlign='left'), 
                        column_selectable='single'
                    )
                ]),
             
        ])
    elif tab == 'Tutorial':
        return html.Div([
            html.H3('Tutorials'),
            html.Div(
                className="tutorial",
                children=[
                ]),
        ])
```
